# Numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp_1 = np.random.uniform(0, 0.5, size=(100, 10, 20, 3))
temp_2 = np.random.uniform(0.5, 1, size=(100, 10, 20, 3))

# (개수, 가로, 세로, 채널(RGB))

print(f'temp_1.shape : {temp_1.shape}')
print(f'temp_2.shape : {temp_2.shape}')

plt.subplot(2,1,1)
plt.axis('off')
plt.title('temp_1[0]')
plt.imshow(temp_1[0])


plt.subplot(2,1,2)
plt.axis('off')
plt.title('temp_2[0]')
plt.imshow(temp_2[0])
```
